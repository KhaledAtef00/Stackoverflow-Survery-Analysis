# Data Preparation

**Execluding Non-Technica Roles**

```python
encoded_df["DevType"].columns
```

```python
# Exclude roles
EXCULDE_ROLES = [
    'Designer',
    'Educator',
    'Marketing or sales professional',
    'Engineering manager',
    'Senior executive/VP',
    'Product manager',
    'Engineer, site reliability']
```

```python
for role in EXCULDE_ROLES:
    encoded_df = encoded_df.drop(("DevType", role), axis=1)
```

```python
encoded_df
```

**Filtering Data**

```python
freq_roles = encoded_df[ROLE_COLS].sum(axis=1)
plt.hist(freq_roles, freq_roles.max())
plt.title('Number of Roles Per Person')
plt.show()
```

```python
freq_tech = encoded_df[TECH_COLS].sum(axis=1)
plt.hist(freq_tech, freq_tech.max())
plt.title('Number of Skills Per Person')
plt.show()
```

```python
# Filter frequencies 
ROLES_RANGE = [1,2]
TECH_RANGE  = [3,12]
```

```python
# Creating masks 
role_mask   = (freq_roles >= ROLES_RANGE[0]) & (freq_roles <= ROLES_RANGE[1])
tech_mask   = (freq_tech >= TECH_RANGE[0])   & (freq_tech <= TECH_RANGE[1])
filter_mask = role_mask & tech_mask
```

```python
features_df = features_df.loc[filter_mask].copy()
encoded_df = encoded_df.loc[filter_mask].copy()
```

```python
encoded_df.shape
```

# Splitting Data 

```python
X = encoded_df.drop("DevType" ,axis=1)
Y = encoded_df["DevType"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y  ,random_state=0)
```

```python
print(X_train.shape, " -- " , X_test.shape)
```

**Balancing Classes In Training Data to Fit Model**

```python
roles_df = Y_train
roles_df.sum()
```

```python
# Resample roles
samples_per_class = 450
resampled_roles = []

for role_col in roles_df.columns:
    sub_df = roles_df.loc[roles_df[role_col] == 1].copy()

    if len(sub_df) < samples_per_class:
        # Upsample
        sub_df = sub_df.sample(samples_per_class, replace=True, random_state=0)
    else:
        # Downsample
        sub_df = sub_df.sample(samples_per_class, random_state=0)

    resampled_roles.append(sub_df)
```

```python
# Construct dfs
Y_train = pd.concat(resampled_roles)
X_train = X_train.loc[Y_train.index].copy()
```

```python
print(X_train.shape , " -- " , X_test.shape)
```

```python
Y_train.sum()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

def calculate_score(Y_pred , Y , metric_function , sort_values=False):
    scores = {}
    for col in Y.columns:
        scores[col] = metric_function(Y[col] , Y_pred[col])*100
        scores[col] = round(scores[col] ,2)
    scores = pd.Series(scores.values() , index = scores.keys())
    if sort_values:
        scores = scores.sort_values()
    return scores
```

```python
model_evaluation = pd.DataFrame(columns = ["Model Name" , "Accuracy_Score" , "Precision_Score" , "Recall_Score", "F1_Score"])
```

```python
X_train.head(10)
```

# Logistic Regression BaseLine Model

```python
from sklearn.pipeline import make_pipeline
from sklearn.multioutput import MultiOutputClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# We Used Multioutput Classifier Because We may assign person with some skills to a different jobs

clf = make_pipeline(StandardScaler(),MultiOutputClassifier(LogisticRegression()))
clf.fit(X_train.values, Y_train.values)
```

```python
#Evaluation On Training Data
Y_pred_train =  pd.DataFrame(clf.predict(X_train.values),columns=Y_train.columns)
train_scores = {score.__name__: calculate_score(Y_pred_train, Y_train, score) 
                for score in [accuracy_score, precision_score, recall_score, f1_score]}
train_scores = pd.concat(train_scores,axis=1)
train_scores
```

```python

#Evaluation On Testing Data
Y_pred_test =  pd.DataFrame(clf.predict(X_test.values),columns=Y_test.columns)
test_scores = {score.__name__: calculate_score(Y_pred_test, Y_test, score) 
                for score in [accuracy_score, precision_score, recall_score, f1_score]}
test_scores = pd.concat(test_scores,axis=1)
test_scores
```

```python
test_scores.mean()
```

```python
test_scores["accuracy_score"].mean()
```

```python
new = pd.DataFrame({
    "Model Name": "BaseLine Logistic Regression",
    "Accuracy_Score": test_scores["accuracy_score"].mean() ,
    "Precision_Score":test_scores["precision_score"].mean(),
    "Recall_Score" :test_scores["recall_score"].mean() , 
    "F1_Score":test_scores["f1_score"].mean(),
}, index=[0])

model_evaluation = model_evaluation.append(new)
```

# Random Forest with Non Linearity
